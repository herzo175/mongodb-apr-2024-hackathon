```python
# Get all frames of video
import tempfile
import glob
import os
from datetime import datetime, timedelta

from showstopper.adapters.ffmpeg import FFMpegAdapter
from showstopper.video.splitting import generate_video_frames

ffmpeg_client = FFMpegAdapter()

frames = generate_video_frames('../inputs/video.mp4', '../outputs/frames', ffmpeg_client)
frames
```

```python
from showstopper.video.splitting import cut_video_shots
from showstopper.adapters.openai import CLIPImageTextEmbedder
# from showstopper.adapters.vit import ViTAdapter

# vit_client = ViTAdapter()
clip_client = CLIPImageTextEmbedder()

shots = cut_video_shots(frames, 0.85, clip_client)
shots
```

```python
# Generate Transcript in SRT format
from showstopper.audio.subtitles import generate_subtitles_from_audio
from showstopper.adapters.openai import OpenAIAdapter

audio_out = "../outputs/audio.mp3"
openai_client = OpenAIAdapter(os.environ["openai_api_key"])

ffmpeg_client.split_audio_video("../inputs/video.mp4", audio_out)

transcript = generate_subtitles_from_audio(audio_out, openai_client)
transcript
```

```python
# from showstopper.video.model import CombinedShot
from showstopper.video.splicing import combine_shot_dialouge_windows

# from datetime import time

# CombinedShot(start_time=time(second=1), end_time=time(second=2), dialouge=None)

merged_windows = combine_shot_dialouge_windows(
    dialouge=transcript,
    shots=shots,
)
merged_windows
```
